```python
df_age_19=pd.read_csv('AgeData/df_age_19.csv',index_col=0)

df_age_19['assists_ev']=df_age_19['assists_ev']/df_age_19['games_played']
df_age_19['assists_pp']=df_age_19['assists_pp']/df_age_19['games_played']
df_age_19['assists_sh']=df_age_19['assists_sh']/df_age_19['games_played']

df_age_19['corsi_for']=df_age_19['corsi_for']/df_age_19['games_played']
df_age_19['corsi_against']=df_age_19['corsi_against']/df_age_19['games_played']

df_age_19['fenwick_for']=df_age_19['fenwick_for']/df_age_19['games_played']
df_age_19['fenwick_against']=df_age_19['fenwick_against']/df_age_19['games_played']

df_age_19['giveaways']=df_age_19['giveaways']/df_age_19['games_played']

df_age_19['goals_ev']=df_age_19['goals_ev']/df_age_19['games_played']
df_age_19['goals_pp']=df_age_19['goals_pp']/df_age_19['games_played']
df_age_19['goals_sh']=df_age_19['goals_sh']/df_age_19['games_played']
df_age_19['goals_ev']=df_age_19['goals_ev']/df_age_19['games_played']

df_age_19['hits']=df_age_19['hits']/df_age_19['games_played']
df_age_19['goals_ev']=df_age_19['goals_ev']/df_age_19['games_played']

df_age_19['pen_min']=df_age_19['pen_min']/df_age_19['games_played']
df_age_19['plus_minus']=df_age_19['plus_minus']/df_age_19['games_played']

df_age_19['shots_attempted']=df_age_19['shots_attempted']/df_age_19['games_played']
df_age_19['takeaways']=df_age_19['takeaways']/df_age_19['games_played']

df_age_19['blocks']=df_age_19['blocks']/df_age_19['games_played']

#df_age_19['time_on_ice_avg'] = df_age_19['time_on_ice_avg'].str.replace(':', '.')
df_age_19
df_age_19=df_age_19.set_index('name')
df_age_19
#df_age_19=df_age_19.convert_objects(convert_numeric=True).dropna()

del df_age_19['TOT']
del df_age_19['season']
del df_age_19['age']
del df_age_19['age_x']

df_age_19=df_age_19.convert_objects(convert_numeric=True).dropna()
df_age_19 =df_age_19.dropna(how='any')
data_age_19=df_age_19.as_matrix()
#df_age_19
#df_age_19.convert_objects(convert_numeric=True).dropna()
#df_age_19.replace([np.inf, -np.inf], np.nan)


pca = decomposition.PCA(n_components=3)
pca.fit(data_age_19)

X = pca.transform(data_age_19)

new_df_age_19 = pd.DataFrame(X)

new_df_age_19['name']=list(df_age_19.index)

new_df_age_19=new_df_age_19.set_index('name')

new_df_age_19.to_csv('PCA_Folder/new_df_age_19_PG.csv')
```
