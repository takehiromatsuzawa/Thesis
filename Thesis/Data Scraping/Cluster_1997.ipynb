```python
results = {}
for i in range(50):
    test_idx = np.random.uniform(0, 1, len(result_all2)) <= 0.7
    train = result_all2[test_idx==True]
    test = result_all2[test_idx==False]
    
    for n in range(1, 15, 1):
        clf = KNeighborsClassifier(n_neighbors=n)
        clf.fit(train[features], train['goals_ev'])
        preds = clf.predict(test[features])
        accuracy = np.where(preds==test['goals_ev'], 1, 0).sum() / float(len(test))
        #print "Neighbors: %d, Accuracy: %3f" % (n, accuracy)
        
        if n in results:
            results[n].append(accuracy)
        else:
            results[n]=[accuracy]
```
