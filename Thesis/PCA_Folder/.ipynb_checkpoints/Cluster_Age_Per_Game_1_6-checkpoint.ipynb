```python
import pandas as pd
import numpy as np
import os
from sklearn import neighbors
from sklearn import cluster, datasets
from sklearn import decomposition
```

```python
for pincipal in range(1,7):
    for age in range(19,39):
        print pincipal, age
        df_age_29=pd.read_csv('../Data/AgeData/age_data/df_age_'+str(age)+'.csv',index_col=0)
        df_age_29=df_age_29[df_age_29['season']!='2016-17']
        df_age_29['assists_ev']=df_age_29['assists_ev']/df_age_29['games_played']
        df_age_29['assists_pp']=df_age_29['assists_pp']/df_age_29['games_played']
        df_age_29['assists_sh']=df_age_29['assists_sh']/df_age_29['games_played']

        df_age_29['corsi_for']=df_age_29['corsi_for']/df_age_29['games_played']
        df_age_29['corsi_against']=df_age_29['corsi_against']/df_age_29['games_played']

        df_age_29['fenwick_for']=df_age_29['fenwick_for']/df_age_29['games_played']
        df_age_29['fenwick_against']=df_age_29['fenwick_against']/df_age_29['games_played']

        df_age_29['giveaways']=df_age_29['giveaways']/df_age_29['games_played']

        df_age_29['goals_ev']=df_age_29['goals_ev']/df_age_29['games_played']
        df_age_29['goals_pp']=df_age_29['goals_pp']/df_age_29['games_played']
        df_age_29['goals_sh']=df_age_29['goals_sh']/df_age_29['games_played']
        df_age_29['goals_ev']=df_age_29['goals_ev']/df_age_29['games_played']

        df_age_29['hits']=df_age_29['hits']/df_age_29['games_played']
        df_age_29['goals_ev']=df_age_29['goals_ev']/df_age_29['games_played']

        df_age_29['pen_min']=df_age_29['pen_min']/df_age_29['games_played']
        df_age_29['plus_minus']=df_age_29['plus_minus']/df_age_29['games_played']

        df_age_29['shots_attempted']=df_age_29['shots_attempted']/df_age_29['games_played']
        df_age_29['takeaways']=df_age_29['takeaways']/df_age_29['games_played']

        df_age_29['blocks']=df_age_29['blocks']/df_age_29['games_played']
        
        df_age_29=df_age_29.set_index('name')

        del df_age_29['TOT']
        del df_age_29['season']
        del df_age_29['uniqueseason']
        del df_age_29['age_x']
        del df_age_29['index_x']
        
        
        cols_to_norm = ['assists_ev','assists_pp','assists_sh', 'blocks','corsi_against','corsi_for', 
        'corsi_pct', 'corsi_rel_pct','fenwick_against', 'fenwick_for','fenwick_pct', 'fenwick_rel_pct',
        'games_played', 'giveaways','goals_ev','goals_pp', 
        'hits', 'on_ice_shot_pct','on_ice_sv_pct',
        'pdo', 'pen_min','plus_minus', 'shot_pct',
        'shots_attempted','takeaways', 'time_on_ice_avg',
         'zs_defense_pct']
    
        df_age_29=df_age_29[cols_to_norm].apply(lambda x: (x - x.mean()) / (x.max() - x.min()))
        
        
        data_age_29=df_age_29.as_matrix()

        pca = decomposition.PCA(n_components=pincipal)
        pca.fit(data_age_29)

        X = pca.transform(data_age_29)

        new_df_age_29 = pd.DataFrame(X)

        new_df_age_29['name']=list(df_age_29.index)

        new_df_age_29=new_df_age_29.set_index('name')

        new_df_age_29.to_csv('PCA_Data/new_df_age_'+str(age)+'_PG_'+str(pincipal)+'.csv')
```
