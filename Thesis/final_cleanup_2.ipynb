```python
import pandas as pd
```

```python
file_name_team='draft_merged_data.csv'
org_data=pd.read_csv(file_name_team,index_col=False)
```

```python
pd.set_option('display.max_columns', None)
del org_data['Unnamed: 0']
```

```python
org_data.head(5)
```

```python
org_data['tm_gm_result_away']=org_data['tm_gm_result_away'].replace(['W'], 1)
org_data['tm_gm_result_away']=org_data['tm_gm_result_away'].replace(['L'], 0)
org_data['tm_gm_result_away']=org_data['tm_gm_result_away'].replace(['T'], 0.5)

org_data['tm_gm_result_home']=org_data['tm_gm_result_home'].replace(['W'], 1)
org_data['tm_gm_result_home']=org_data['tm_gm_result_home'].replace(['L'], 0)
org_data['tm_gm_result_home']=org_data['tm_gm_result_home'].replace(['T'], 0.5)

org_data['tm_gm_loc_away']=org_data['tm_gm_loc_away'].replace(['H'], 1)
org_data['tm_gm_loc_away']=org_data['tm_gm_loc_away'].replace(['A'], 0)

org_data['tm_gm_loc_home']=org_data['tm_gm_loc_home'].replace(['H'], 1)
org_data['tm_gm_loc_home']=org_data['tm_gm_loc_home'].replace(['A'], 0)
```

```python
cum_results=org_data['tm_loss_away'].values

new_cum_result=[]
for cum_result in cum_results:
    if cum_result[0]=='T':
        new_cum_result.append(0)
    elif cum_result[0]=='W':
        new_cum_result.append(int(cum_result[2:]))
    elif cum_result[0]=='L':
        new_cum_result.append((-1)*int(cum_result[2:]))

org_data['tm_loss_away']=new_cum_result
```

```python
cum_results=org_data['tm_loss_home'].values

new_cum_result=[]
for cum_result in cum_results:
    if cum_result[0]=='T':
        new_cum_result.append(0)
    elif cum_result[0]=='W':
        new_cum_result.append(int(cum_result[2:]))
    elif cum_result[0]=='L':
        new_cum_result.append((-1)*int(cum_result[2:]))

org_data['tm_loss_home']=new_cum_result
```

```python
org_data['tm_num_games_low']=org_data['tm_num_games_home']-20
org_data['tm_num_games_high']=org_data['tm_num_games_home']+10
```

```python
frames = []
data_len=org_data.shape[0]
for nth_game in range(0,data_len):
# for nth_game in range(0,100):
    this_data=org_data.iloc[nth_game]
    url_fltr,year_fltr,curr_game,game_min,game_max=this_data['Url'],this_data['Year'],this_data['tm_num_games_home'],this_data['tm_num_games_low'],this_data['tm_num_games_high']
    small_data=org_data[(org_data['Url']==url_fltr) & (org_data['Year']==year_fltr) & (org_data['player_num_games']>game_min)& (org_data['player_num_games']<curr_game)]
    predict_data=org_data[(org_data['Url']==url_fltr) & (org_data['Year']==year_fltr) & (org_data['player_num_games']>curr_game)& (org_data['player_num_games']<game_max)]

    
    

    if small_data.shape[0]>=10 and predict_data.shape[0]>=5:
        small_data=small_data.iloc[-10:]
        ShotHand=small_data.iloc[0]['ShotHand']
        Position=small_data.iloc[0]['Position']
        Url=small_data.iloc[0]['Url']
        Name=small_data.iloc[0]['Name']

        Height=small_data.iloc[0]['Height']
        Weight=small_data.iloc[0]['Weight']
        Drafts=small_data.iloc[0]['Drafts']
        draft_years=small_data.iloc[0]['draft_years']
        Year=small_data.iloc[0]['Year']

        predict_data=predict_data[:5]
        pdt_goal_all=predict_data['goals_all'].mean()
        pdt_asst_all=predict_data['assists_all'].mean()

        mean_row=small_data[['player_gm_result','age','player_num_games','player_gm_loc','goals_all','assists_all','plus_minus','pen_min','goals_ev',\
                            'goals_pp', 'goals_sh', 'assists_pp','assists_ev', 'assists_sh', 'num_shifts', 'time_on_ice',\
                           'hits_all', 'blocks_all', 'fc_win', 'fc_loss','cum_loss_for_home','cum_loss_for_away',\
                           'cum_wins_for_home','cum_wins_for_away','opp_pen_min_away','opp_pen_min_home','opp_pen_min_away',\
                            'tm_goals_agt_pp_away','tm_goals_agt_pp_home', 'tm_goals_agt_sh_away','tm_goals_agt_sh_home', \
                             'tm_goals_for_away', 'tm_goals_for_home','tm_goals_pp_for_away', 'tm_goals_pp_for_home',\
                            'tm_goals_sh_for_away', 'tm_goals_sh_for_home', 'tm_loss_away','tm_loss_home', 'tm_num_games_away',\
                             'tm_num_games_home','tm_pen_min_away', 'tm_pen_min_home', 'tm_shots_agt_away',\
                            'tm_shots_agt_home', 'tm_shots_for_away', 'tm_shots_for_home']].mean()

        new_df=mean_row.to_frame().T
        new_df['ShotHand']=ShotHand
        new_df['Position']=Position
        new_df['Url']=Url
        new_df['Name']=Name

        new_df['Year']=Year

        new_df['Height']=Height
        new_df['Weight']=Weight
        new_df['Drafts']=Drafts
        new_df['response_assists']=pdt_asst_all
        new_df['response_goal']=pdt_goal_all
        new_df['response_points']=pdt_asst_all+pdt_goal_all
        
        frames.append(new_df)
```

```python
new_data=pd.concat(frames,ignore_index=True)
```
