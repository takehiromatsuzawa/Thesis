```python
def continuous(array):
    curr_max=array[0]
    total_max=array[0]
    for i in range(1,len(array)):
        curr_max=max(curr_max+array[i],array[i])
        total_max=max(total_max,curr_max)
        
    return total_max
```

```python
a=[-2, -3, 4, -1, -2, 1, 5, -3]
continuous(a)
```

```python
def search_rotate(array,high,low,target):
    if low>high:
        return
    mid=(low+high)//2
    if array[mid]==target:
        print mid
        return mid
    elif array[low]<array[mid]:
        if array[low]<target and target<array[mid]:
            return search_rotate(array,mid-1,low,target)
        else:
            return search_rotate(array,high,mid+1,target)
    else:
        if array[mid]<target and target<array[high]:
            return search_rotate(array,high,mid+1,target)
        else:
            return search_rotate(array,mid-1,low,target)
        
```

```python
array=[3, 4, 5, 1, 2]

search_rotate(array,4,0,4)
```

```python
def max_sum(array):
    inc=array[0]
    exc=0
    for i in range(1,len(array)):
        excl_new=max(inc,exc)
        inc=exc+array[i]
        exc=excl_new
        
    return max(inc,exc)
```

```python
max_sum([5,  5, 10, 40, 50, 35])
```

```python
def seg_even_odd(array):
    end=len(array)-1
    start=0
    while start<end:
        if array[start]%2==1 and start<end:
            start=start+1
        elif array[end]%2==0 and start<end:
            end=end-1
        elif start<end:
            array[start],array[end]=array[end],array[start]
            start=start+1
            end=end-1
```

```python
alist=[5,  5, 10, 40, 50, 35]
seg_even_odd(alist)
alist
```

```python
def find_top(array,low,high):
    #if low>high:
    #    return array[high]
    
    if low==high:
        return array[low]
    elif low==high-1:
        return max(array[low],array[high])
    else:
        mid=(low+high)//2
        if array[mid]>array[mid-1] and array[mid]>array[mid+1]:
            return array[mid]
        else:
            if array[mid]>array[mid-1]:
                return find_top(array,mid+1,high)
            else:
                return find_top(array,low,mid-1)
```

```python
alist=[8, 10, 20, 80, 100, 200, 400, 500,3,2,1]
find_top(alist,0,len(alist)-1)
```

```python
def peak(array,low,high):
    if low==high:
        return array[low]
    elif low==high-1:
        return max(array[low],array[high])
    else:
        mid=(low+high)//2
        if array[mid-1]<array[mid] and array[mid]>array[mid+1]:
            return mid
        elif array[mid-1]<array[mid]:
            return peak(array,mid+1,high)
        else:
            return peak(array,low,mid-1)
```

```python
alist=[500,400,3,2,1]
peak(alist,0,len(alist)-1)
```

```python
def mini(array,low,high):
    if low==high:
        return array[low]
    elif low==high-1:
        return min(array[low],array[high])
    else:
        mid=(low+high)//2
        
        if array[mid-1]>array[mid] and array[mid]<array[mid+1]:
            return mid
        
        elif array[mid]>array[low] and array[low]>array[high]:
            return peak(array,mid+1,high)
        
        else:
            return peak(array,low,mid-1)
        
        
```
