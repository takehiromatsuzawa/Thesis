```python
web='https://en.wikipedia.org/wiki/List_of_American_and_Canadian_cities_by_number_of_major_professional_sports_franchises#cite_note-CFL_teams-8'
cap = requests.get(web)
capText = BeautifulSoup(cap.text)
resultCap= capText.find_all("table",{'class':"wikitable sortable"})[1]

p = resultCap.find_all('tr')
paragraphs = []
for num in range(len(p)):
    if num==0:
        elem=p[num].find_all('th')
    else:
        elem=p[num].find_all('td')
    result=[]
    for elem1 in elem:
        result.append(elem1)
    paragraphs.append(result)

import re
CLEAN = re.compile('\s+')
newpar=[]
for elem in paragraphs:
    elem = [CLEAN.sub(' ', a.text).strip() for a in elem]
    newpar.append(elem)

numCity = pd.DataFrame(newpar)
numCity.columns = numCity.iloc[0]
numCity=numCity.drop(numCity.index[0])

columnsTeam=[numCity.columns[0],numCity.columns[3],numCity.columns[4],numCity.columns[5],numCity.columns[6]]
dfCity = pd.DataFrame(numCity,index=numCity.index, columns=columnsTeam)

dfCity[dfCity.columns[1]] = dfCity[dfCity.columns[1]].str[:1]
dfCity[dfCity.columns[2]] = dfCity[dfCity.columns[2]].str[:1]
dfCity[dfCity.columns[3]] = dfCity[dfCity.columns[3]].str[:1]
dfCity[dfCity.columns[4]] = dfCity[dfCity.columns[4]].str[:1]
dfCity=dfCity.drop(dfCity.index[len(dfCity)-1])

dfCity.columns = ['State','NFL','MLB','NBA','NHL']
dfCity
```

```python
cap = requests.get("http://war-on-ice.com/cap/")
capText = BeautifulSoup(cap.text)
resultCap= capText.find_all("a")
link=[]
for elem in resultCap:
    if elem.has_attr('href'):
        link.append(elem.attrs['href'])
linkList=link[6:36]

NHLSalary = pd.DataFrame()
for links in linkList:
    r3 = requests.get(links)
    text3 = r3.text
    NHLsalary_data=pd.read_html(text3)
    NHL_salary=NHLsalary_data[0].copy()
    NHL_salary=NHL_salary.drop(NHL_salary.index[0])
    NHL_salary
    NHL_salary.columns = NHL_salary.iloc[0]
    NHL_salary['team']=links[-8:-5]
    NHL_salary.reset_index(drop=True)
    NHLSalary =NHLSalary.append(NHL_salary)

NHLSalary=NHLSalary.dropna()
NHLSalary=NHLSalary[NHLSalary['2015-16']!='--']
NHLSalary=NHLSalary[NHLSalary['2015-16']!='UFA']
NHLSalary=NHLSalary[NHLSalary['2015-16']!='RFA']
NHLSalary=NHLSalary[NHLSalary['2015-16']!='2015-16']
del NHLSalary[NHLSalary.columns[2]]
del NHLSalary[NHLSalary.columns[2]]
del NHLSalary[NHLSalary.columns[3]]
del NHLSalary[NHLSalary.columns[3]]
del NHLSalary[NHLSalary.columns[3]]
del NHLSalary[NHLSalary.columns[3]]

NHLSalary['2015-16'] = NHLSalary['2015-16'].str.replace(r'[$,]', '').astype('float')

NHLSalary
NHLSalary.columns = [ 'Player', 'Pos', 'Salary', 'Team']
NHLSalary=NHLSalary.reset_index(drop=True)
NHLSalary.head()
```
