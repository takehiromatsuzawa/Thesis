```python
import unicodedata
import re

# -*- coding: utf-8 -*-
years=range(1992, 2015)
listClean=[]

for year in years:
    url="http://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_%s" %year
    t=requests.get(url).text.encode('utf-8')
    soup = BeautifulSoup(t, "html.parser")


    rows = soup.find("table", attrs={"class": "wikitable"}).find_all("tr")[1:]

    cleaner = lambda r: [ r[0].get_text(), r[1].findAll("a"), r[1].findAll("a"),r[0].findAll("a"),r[0].findAll("a")]

    cleaned=[]

    for row in rows:
        #         get the data of all the td and apply cleaner function
        clearnerd=cleaner(row.find_all("td"))
        listHref=[]
        listSinger=[]
        listSongs=[]
        #         listHref has all the urls of singer's websites
        if len(clearnerd[2])==0:
            listHref.append(None)
            listSinger.append(None)
        else:
            for elem in clearnerd[2]:
                listHref.append(str(elem.get('href')))
#                 change the nonascii to ascii
                singerName=str(unicodedata.normalize('NFKD', elem.get('title')).encode('ascii', 'ignore'))
#                 print singerName
#                 if there is (), I remove them because they only mean kinds of singers such as band, group and etc
                if '(' in singerName:
                    singerName=singerName.rsplit('(', 1)[0][:len(singerName.rsplit('(', 1)[0])-1]
                listSinger.append(singerName)
        #         listHref has all the urls of song's urls
        if len(clearnerd[4])==0:
            listSongs.append(None)
        else:
            for elem in clearnerd[4]:
                #                 change the nonascii to ascii
                songName=str(unicodedata.normalize('NFKD', elem.get('title')).encode('ascii', 'ignore'))
                #                 if there is (), I remove them because they only mean kinds of music
                if '(' in songName:
                    songName=songName.rsplit('(', 1)[0][:len(songName.rsplit('(', 1)[0])-1]
                listSongs.append(songName)
        clearnerd[2]=listHref
        clearnerd[4]=listSongs
        clearnerd[1]=listSinger
         #         listsongHref has all the urls of song's websites
        listsongHref=[]
        if len(clearnerd[3])==0:
            listsongHref.append(None)
        else:
            for elem in clearnerd[3]:
                listsongHref.append(str(elem.get('href')))
        
        clearnerd[3]=listsongHref
        
        cleaned.append(clearnerd)
        
#     insert the year and ranking into cleaned
    for i in range(0,len(cleaned)):
        cleaned[i].insert(0,i+1)
        cleaned[i].insert(0,year)

    listClean.append(cleaned)
# fields are keys of a new dictionary
fields = [ "year","ranking","titletext", "band_singer", "url","songurl","song"]   
# Since listClean is a triple list, I have to convert it into a double list(newlistClean)
newlistClean=[]
for i in range(0,len(listClean)):
    newlistClean=newlistClean+listClean[i]
# make a dictionary
yearstext = [dict(zip(fields, row)) for row in newlistClean]

```

```python
"""
Function
--------
parse_year

Inputs
------
year: the year you want the singles for
ytextdixt: a dictionary with keys as integer years and values the downloaded web pages 
    from wikipedia for that year.
   
Returns
-------

a list of dictionaries, each of which corresponds to a single and has the
following data:

Eg:

{'band_singer': ['Brandy', 'Monica'],
  'ranking': 2,
  'song': ['The Boy Is Mine'],
  'songurl': ['/wiki/The_Boy_Is_Mine_(song)'],
  'titletext': '" The Boy Is Mine "',
  'url': ['/wiki/Brandy_Norwood', '/wiki/Monica_(entertainer)']}
 
A dictionary with the following data:
    band_singer: a list of bands/singers who made this single
    song: a list of the titles of songs on this single
    songurl: a list of the same size as song which has urls for the songs on the single 
        (see point 3 above)
    ranking: ranking of the single
    titletext: the contents of the table cell
    band_singer: a list of bands or singers on this single
    url: a list of wikipedia singer/band urls on this single: only put in the part 
        of the url from /wiki onwards
    

Notes
-----
See description and example above.
"""

#your code here
import copy
import re
def parse_year(year,ytextdixt):
     
#     copy ytextdixt
    temp1 = copy.deepcopy(ytextdixt)
# I do not have to care about ranking and year.
    dictElem=fields[2:]

    for elem in temp1 :
        for dictElem1 in dictElem: 
#             if the band_singer type is string, make it a list
            if dictElem1=="band_singer" and type(elem[dictElem1])==str:
                elem[dictElem1]=[elem[dictElem1]]   
#             if the song type is string, make it a list
            if dictElem1=="song" and type(elem[dictElem1])==str:
                elem[dictElem1]=[elem[dictElem1]]
#     drop the other years
    temp=[elem for elem in temp1 if elem["year"]==year]

    for elem1 in temp:
        del elem1["year"]

    return temp
    

 
```
